```python
import pandas as pd
from glob import glob
np = pd.np
```

```python
def remove_location_from_header(row):
#     print(row)
    text = row['text']
    location = row['location']
    if text.startswith(location):
        text = text[len(location) + 2:]
    row['text'] = text.strip()
    return row
```

```python
df_list = []
for file in glob("../../data/TOI/*"):
    temp_df = pd.read_csv(file, index_col=0)
    # Get the date from the file name
    temp_df['date'] = pd.to_datetime(file.split("/")[-1].split(".")[0])
    df_list.append(temp_df.apply(remove_location_from_header, axis=1))
df = pd.concat(df_list)
```

```python
df = df.sort_values(by=["date"])
df['location'] = df['location'].apply(str.title)
df.index = np.arange(len(df))
```

```python
from geotext import GeoText
```

```python
def is_valid_location(row):
    places =  GeoText(row.name)
    if places.cities:
        row.name = places.cities
    else:
        return None
```

```python
valid_location = df.groupby('location').count().apply(is_valid_location, axis=1)
```
